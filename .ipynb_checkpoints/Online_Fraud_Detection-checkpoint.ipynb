```python
# importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing dataset
data = pd.read_csv("D:\GitHub Contributions\Datasets\Online_Fraud_Detection_Dataset_log.csv")
data
```

```python
data.info()
```

```python
data = data[['amount','oldbalanceOrg','newbalanceOrig','oldbalanceDest','newbalanceDest',"isFraud","isFlaggedFraud"]]
data.info()
```

```python
data.isnull().sum()
```

```python
data["isFraud"].value_counts()
```

```python
X = data.drop('isFraud', axis = 1)
y = data['isFraud']
data.info()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 30, random_state = 1)
```

```python
X_train
```

```python
# Train the Logistic Regression Model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression (solver='liblinear')
classifier.fit(X_train,y_train)
```

```python
# Prediction with the test set
y_predict = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_predict))
```

```python
y_train_pred = classifier.predict(X_train)
print(accuracy_score(y_train,y_train_pred))
```

```python
# Getting probability predictions from the model.
y_test_proba = classifier.predict_proba(X_test)
print(y_test_proba.shape)
```
